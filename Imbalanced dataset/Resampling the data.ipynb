```python
targetcount = train.target.value_counts()

#use print statement to count the number of targets of 0 and 1
print("Class 0", targetcount[0])
print("Class 1", targetcount[1])
print("Proportion of Class 0 to class 1: ", round(targetcount[0]/targetcount[1],2),":1")

targetcount.plot(kind="bar",title='Count of title')
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

labels = train.columns[2:]
x = train[labels]
y = train['target']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
model = XGBClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
